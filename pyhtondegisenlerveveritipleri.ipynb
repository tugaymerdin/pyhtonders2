```python
# Dictionary - Sözlük Örneği -> öğrencilerin notlarını tutan sozluk
notlar= {
    "tugay": 100,
    "akbas": 90,
    "omer": 80
}
notlar["fatma"] = 92 #yeni ogrenci ekleme
notlar["tugay"] = 88 #bir ogrencinin notunu guncelleme

print(notlar)
```

```python
print(notlar.keys()) # dict_keys (['isim','yas','meslek'])

print(notlar.values()) #dict_values(['ahmet',26, 'muhendis])
print(notlar.items()) #dict_items ([('isim','ahmet')('yas',26),('meslek', 'muhendis)])
```

```python
#Set - Küme örnegi --> bir yazilim etkinligine katilan kisilerin olusturudgu kumeler
kodyazanlar = {"tugay","akbas","omer"}
kodyazmayanlar = {"tugay","a","b"}

#kafasina gore takılan ogrenciler
ortak= kodyazanlar.intersection(kodyazmayanlar)
print("kafasina gore calisan ogrenciler:", ortak)

#etkinlikte kodyazmayan ogrencilerin tespiti
kod_yazmayan = kodyazmayanlar.difference(kodyazanlar)
print("kod yazmayanlar:", kod_yazmayan)

# yazılım egitiminin ogrenci imza listesi
imza_listesi = kodyazanlar.union(kodyazmayanlar)
print("ogrenci imza listesi:", imza_listesi)

```
